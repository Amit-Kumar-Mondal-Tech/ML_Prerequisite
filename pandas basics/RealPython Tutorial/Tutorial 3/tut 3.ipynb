## Pandas Sort: Your Guide to Sorting Data in Python

```python
import pandas as pd
import numpy as np
```

```python
column_subset = [
...     "id",
...     "make",
...     "model",
...     "year",
...     "cylinders",
...     "fuelType",
...     "trany",
...     "mpgData",
...     "city08",
...     "highway08"
... ]
```

```python
df = pd.read_csv(
...     "data/vehicles.csv",
...     usecols=column_subset,
...     nrows=100
... )
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df = df.astype(dtype={'city08':'uint8', 
                       'cylinders':'uint8',
                       'highway08':'uint8',
                       'id':'uint16'})
```

```python
df.info(0)
```

```python
df['fuelType'].unique()
```

```python
df = df.astype(dtype={'year':'uint16', 'mpgData':'category', 'fuelType':'category'})
```

```python
df.info()
```

```python
df.sort_values('city08', ascending=False)
```

```python
cars = df[['make', 'model']].groupby('make')
cars.get_group('Audi')
```

```python
cars
```

```python
assigned_index_df = df.set_index(
...     ["make", "model"])
```

```python
assigned_index_df
```

```python
df.sort_index(axis=1, ascending=False)
```

```python
df.sort_index(axis=0, ascending=False)
```

```python
test_col = df['mpgData'].apply(lambda res: True if res == 'Y' else np.nan)
```

```python
test_col
```
